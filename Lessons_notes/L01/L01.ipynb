### Gradient descent
A simple first approach is to initialize the variable to a random number and then descent down the slope, as seen in other courses, it depend on the value, the learning rate and the gradient.

### Loss function
Needed to compare the output of our model to the ground truth, to estimate the goodness of the estimations we did. It drives the backpropagation.  

### Forward propagation
Before coming to the loss function we do a forward propagation to estimate our value, it goes trough multiple functions, which is exactly the main characteristic of deep learning.

### Backward propagation
It's needed to update the weights of our model.

### Model fitting:

####  - Initialization

Different initialization impacts the final model we get, the various initializatons won't always converge to the same model.

#### - Learning rate

if we use a learning rate too high, it could never converge, the opnly way to understand if we're doing it right we have to analize the loss function.  
if we use a learning rate too low, it may converge in not enough time for us to be alive.


### Common problems:

- Overfitting, too complex model
- Underfitting, too simple model
